```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Load the dataset

```python
df = pd.read_csv('Electric_Vehicle_Population_Data_20260126.csv')
sns.set(style="whitegrid")
```

Q1: How has the number of registered EVs grown over the years?

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.groupby('Model Year').size())
plt.title('Q1: Growth of EV Registrations by Model Year')
plt.ylabel('Number of Vehicles')
plt.savefig('q1_growth.png')
plt.show()
```

Q2: What is the distribution of Electric Vehicle Types (BEV vs PHEV)?

```python
plt.figure(figsize=(8, 6))
df['Electric Vehicle Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=['#66b3ff','#99ff99'])
plt.title('Q2: Distribution of Electric Vehicle Types')
plt.ylabel('')
plt.savefig('q2_types.png')
plt.show()
```

Q3: Who are the top 10 EV manufacturers by market share?

```python
plt.figure(figsize=(12, 6))
top_makes = df['Make'].value_counts().head(10)
sns.barplot(x=top_makes.values, y=top_makes.index, palette='viridis')
plt.title('Q3: Top 10 EV Manufacturers')
plt.savefig('q3_top_makes.png')
plt.show()
```

Q4: What are the most popular EV models currently registered?

```python
plt.figure(figsize=(12, 6))
top_models = df['Model'].value_counts().head(10)
sns.barplot(x=top_models.values, y=top_models.index, palette='magma')
plt.title('Q4: Top 10 Popular EV Models')
plt.savefig('q4_top_models.png')
plt.show()
```

Q5: Which counties have the highest EV adoption?

```python
plt.figure(figsize=(12, 6))
top_counties = df['County'].value_counts().head(10)
sns.barplot(x=top_counties.values, y=top_counties.index, palette='coolwarm')
plt.title('Q5: Top 10 Counties by EV Adoption')
plt.savefig('q5_top_counties.png')
plt.show()
```

Q6: What are the top cities for EV ownership?

```python
plt.figure(figsize=(12, 6))
top_cities = df['City'].value_counts().head(10)
sns.barplot(x=top_cities.values, y=top_cities.index, palette='cubehelix')
plt.title('Q6: Top 10 Cities for EV Ownership')
plt.savefig('q6_top_cities.png')
plt.show()
```

Q7: How are EVs distributed across Legislative Districts?

```python
plt.figure(figsize=(14, 6))
district_counts = df['Legislative District'].value_counts().sort_index()
sns.barplot(x=district_counts.index.astype(int), y=district_counts.values, color='skyblue')
plt.title('Q7: EV Distribution by Legislative District')
plt.xticks(rotation=90)
plt.savefig('q7_districts.png')
plt.show()
```

Q8: What is the distribution of Electric Range for BEVs vs PHEVs?

```python
df_range = df[df['Electric Range'] > 0]
plt.figure(figsize=(10, 6))
sns.histplot(data=df_range, x='Electric Range', hue='Electric Vehicle Type', kde=True, bins=30, palette='Set2')
plt.title('Q8: Distribution of Electric Range (Non-Zero)')
plt.savefig('q8_range_dist.png')
plt.show()
```

Q9: Top 10 EV Makes by Count

```python
top_makes = df['Make'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=top_makes.values, y=top_makes.index, palette='viridis')
plt.title('Top 10 EV Makes by Registration Count')
plt.xlabel('Number of EVs')
plt.tight_layout()
plt.show()
```

Q10: EVs by Electric Range (Histogram)

```python
plt.figure(figsize=(10,6))
plt.hist(df['Electric Range'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Frequency')
plt.axvline(df['Electric Range'].mean(), color='red', linestyle='--', label='Mean')
plt.legend()
plt.show()
```

Q11: Base MSRP vs Electric Range (Scatter)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Electric Range', y='Model Year', hue='Electric Vehicle Type', alpha=0.6)
plt.title('Electric Range vs Model Year by EV Type')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Model Year')
plt.tight_layout()
plt.show()
```

Q12: EVs by County (Top 10 Heatmap)

```python
county_counts = df['County'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=county_counts.values, y=county_counts.index, palette='plasma')
plt.title('Top 10 Counties by EV Registrations')
plt.xlabel('Number of EVs')
plt.show()
```

Q13: Model Year vs Registration Growth (Line + Bar)

```python
year_counts = df['Model Year'].value_counts().sort_index()
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,6))
year_counts.plot(kind='line', ax=ax1, marker='o')
ax1.set_title('EV Registrations by Model Year')
ax1.set_ylabel('Count')

sns.countplot(data=df, x='Model Year', ax=ax2, palette='coolwarm')
ax2.tick_params(axis='x', rotation=45)
ax2.set_title('EV Registrations by Model Year (Bar)')
plt.tight_layout()
plt.show()
```

Q14: EV Type Proportions (Pie)

```python
type_counts = df['Electric Vehicle Type'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(type_counts.values, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('EV Types Distribution')
plt.show()
```

Q15: Correlation Heatmap (Key Features)

```python
# Actual numeric columns in your dataset
numeric_cols = ['Model Year', 'Electric Range']

corr = df[numeric_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Q15: Correlation Matrix - Model Year vs Electric Range')
plt.tight_layout()
plt.savefig('q15_correlation.png')
plt.show()
```
